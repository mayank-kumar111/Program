```python
'''1. plt.plot()
Definition: Used to create a line plot by connecting data points with straight lines.

Syntax: plt.plot(x, y, format_string, **kwargs)

Example:'''
import matplotlib.pyplot as plt
'''x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, 'bo-')  # 'bo-' means blue color, circle markers, solid line
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()'''
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
'''2. plt.xlabel() and plt.ylabel()
Definition: Sets the labels for the x-axis and y-axis respectively.

Syntax: plt.xlabel("text") | plt.ylabel("text")

Example:'''
plt.plot([1, 2], [10, 20])
plt.xlabel("Time (s)")
plt.ylabel("Speed (m/s)")
plt.show()
```

```python
'''3. plt.title()
Definition: Adds a title to the top of the current plot.

Syntax: plt.title("Title Name")

Example:'''
plt.plot([1, 2], [5, 5])
plt.title("Constant Velocity Graph")
plt.show()
```

```python
'''4. plt.scatter()
Definition: Creates a scatter plot where individual data points are represented as dots without connecting lines.

Syntax: plt.scatter(x, y, color, marker)

Example:'''
plt.scatter([1, 2, 3, 4], [10, 15, 7, 12], color='red')
plt.show()
```

```python
'''5. plt.bar()
Definition: Creates a vertical bar chart to compare categories of data.

Syntax: plt.bar(x, height, width, color)

Example:'''
plt.bar(['A', 'B', 'C'], [10, 20, 15], color='green')
plt.show()
```

```python
'''6. plt.hist()
Definition: Computes and draws a histogram to show the frequency distribution of a dataset.

Syntax: plt.hist(data, bins)

Example:'''

data = [1, 2, 2, 3, 3, 3, 4, 5]
plt.hist(data, bins=5)
plt.show()
```

```python
'''7. plt.pie()
Definition: Creates a circular pie chart to show proportions of a whole.

Syntax: plt.pie(sizes, labels, autopct)

Example:'''

plt.pie([30, 70], labels=['Task A', 'Task B'], autopct='%1.1f%%')
plt.show()
```

```python
'''8. plt.legend()
Definition: Adds a legend (key) to identify different lines or data series in a plot.

Syntax: plt.legend(loc="upper right")

Example:'''

plt.plot([1, 2], [1, 2], label='Growth')
plt.legend()
plt.show()
```

```python
'''9. plt.grid()
Definition: Adds grid lines (vertical and horizontal) to the background of the plot for easier reading.

Syntax: plt.grid(True)

Example:'''

plt.plot([1, 5], [1, 5])
plt.grid(True)
plt.show()
```

```python
'''10. plt.savefig()
Definition: Saves the current figure to a file (PNG, JPG, PDF, etc.) instead of just displaying it.

Syntax: plt.savefig("filename.png")

Example:'''

plt.plot([1, 2, 3], [1, 2, 3])
plt.savefig("my_plot.png")
```

```python
'''Category: Advanced Plotting
1. plt.stem()
Definition: Creates a stem plot, which is useful for visualizing discrete data points (like digital signals).

Syntax: plt.stem(x, y)

Example:'''

x = [0, 1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 2, 6]
plt.stem(x, y)
plt.show()
```

```python
'''2. plt.boxplot()
Definition: Draws a box-and-whisker plot to show the distribution of data based on a five-number summary (Minimum, First Quartile, Median, Third Quartile, and Maximum).

Syntax: plt.boxplot(data)

Example:'''

data = [10, 20, 22, 25, 30, 35, 50]
plt.boxplot(data)
plt.show()
```

```python
'''Category: Visual Customization
3. plt.grid()
Definition: Configures the visibility and appearance of grid lines in the background.

Syntax: plt.grid(color, linestyle, linewidth)

Example:'''

plt.plot([1, 2, 3], [10, 20, 30])
plt.grid(True, color='grey', linestyle='--')
plt.show()
```

```python
'''4. plt.axis()
Definition: Allows you to set the range of the x and y axes or change the scale (e.g., equal scaling).
Syntax: plt.axis([xmin, xmax, ymin, ymax]) or plt.axis('equal')
Example:'''

plt.plot([1, 2, 3], [1, 2, 3])
plt.axis([0, 5, 0, 10])
plt.show()
```

```python
'''5. plt.text()
Definition: Adds a text string at an arbitrary $(x, y)$ location on the plot.
Syntax: plt.text(x, y, "string")
Example:'''

plt.plot([1, 2, 3], [1, 4, 9])
plt.text(2, 4, "Peak Point")
plt.show()
```

```python
'''Category: Figure & Coordinate Management
6. plt.figure()
Definition: Used to create a new figure object or adjust the size of the plot.

Syntax: plt.figure(figsize=(width, height))

Example:'''

plt.figure(figsize=(10, 5))
plt.plot([1, 2], [1, 2])
plt.show()
```

```python
'''7. plt.xticks() and plt.yticks()
Definition: Sets the tick locations and labels on the x or y axis.

Syntax: plt.xticks(ticks, labels)

Example:'''

plt.plot([1, 2, 3], [10, 20, 30])
plt.xticks([1, 2, 3], ['Low', 'Med', 'High'])
plt.show()
```

```python
'''8. plt.tight_layout()
Definition: Automatically adjusts subplot parameters so that subplots fit into the figure area without overlapping.

Syntax: plt.tight_layout()

Example:'''

plt.subplot(2, 1, 1)
plt.title("Title 1")
plt.subplot(2, 1, 2)
plt.title("Title 2")
plt.tight_layout()
plt.show()
```

```python
'''Category: Specialized Visuals
9. plt.errorbar()
Definition: Plots data points with attached error bars to show uncertainty or variability.

Syntax: plt.errorbar(x, y, yerr=error_values)

Example:'''

x = [1, 2, 3]
y = [10, 20, 30]
error = [1, 2, 1.5]
plt.errorbar(x, y, yerr=error, fmt='o')
plt.show()
```

```python
'''10. plt.fill_between()
Definition: Fills the area between two horizontal curves with a specific color.

Syntax: plt.fill_between(x, y1, y2, color)

Example:'''

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y)
plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.show()
```

```python
Matplotlib in Python
1
2
3
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It was created by John D. Hunter and is now maintained by a large community of developers. Matplotlib is widely used for data visualization in various fields, including scientific research, finance, data analysis, education, and engineering.

Key Features of Matplotlib

Versatility: Matplotlib can generate a wide range of plots, including line plots, scatter plots, bar plots, histograms, pie charts, and more.

Customization: It offers extensive customization options to control every aspect of the plot, such as line styles, colors, markers, labels, and annotations.

Integration with NumPy: Matplotlib integrates seamlessly with NumPy, making it easy to plot data arrays directly.

Publication Quality: Matplotlib produces high-quality plots suitable for publication with fine-grained control over the plot aesthetics.

Interactive Plots: Matplotlib supports interactive plotting through the use of widgets and event handling, enabling users to explore data dynamically.

Basic Example

Here is a basic example of how to create a simple line plot using Matplotlib:

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y)

# Add title and labels
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
Copy
Components of a Matplotlib Figure

Figure: The top-level container that holds all the elements of a plot.

Axes: The rectangular areas within the figure where data is plotted.

Axis: Represents the x-axis and y-axis of the plot.

Markers: Symbols used to denote individual data points on a plot.

Lines: Connect data points on a plot and represent the relationship or trend between data points.

Title: A text element that provides a descriptive title for the plot.

Labels: Text elements that provide descriptions for the x-axis and y-axis.

Ticks and Tick Labels: Small marks along the axis that indicate specific data points or intervals.

Legend: Provides a key to the symbols or colors used in the plot.

Grid Lines: Horizontal and vertical lines that extend across the plot.

Spines: The lines that form the borders of the plot area.

Advantages and Disadvantages

Advantages

Versatility: Can create a wide range of plots.

Customization: Extensive options to control every aspect of the plot.

Integration with NumPy: Easy to plot data arrays directly.

Publication Quality: Produces high-quality plots.

Interactive Plots: Supports interactive plotting.

Disadvantages

Steep Learning Curve: Can be challenging for beginners.

Verbose Syntax: Can be time-consuming to create and customize plots.

Limited Interactivity: Does not offer as many interactive features as other libraries.

Performance Issues with Large Datasets: Can be slower and less efficient when plotting large datasets.

Matplotlib remains a popular choice for data visualization in Python due to its flexibility, versatility, and extensive capabilities.
```
