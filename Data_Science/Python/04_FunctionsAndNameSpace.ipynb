##### Filter()

```python
lst = [23,45,67,789,45,89,56,89,56,90,45,89] 
```

```python
def even_odd(num):
    if num % 2 == 0:
        return True
    else:
        return False
```

```python
for i in lst:
    if even_odd(i):
        print(i)
```

```python
lst
```

```python
list(filter(even_odd,lst))
```

```python
list(filter(lambda x: True if x%2 == 0 else False,lst))
```

```python
names
```

```python
list(filter(lambda x: True if len(x) > 5 else False, names))
```

```python
num = int(input("Enter num: "))

fact = 1

for i in range(1, num+1):
    fact = fact * i
else:
    print(fact)
```

```python
num = int(input("Enter num: "))

fact = 1

i = 1
while i < num+1:
    fact *= i
    i += 1
else:
    print(fact)
```

#### Namespaces

    global variables:
                        variable declare outside functions are called global variable.
                        we can use global variable anywhere in the program
                        we can not change global variable directly in local space
    local variables:
                    variables declare inside the functions are called local variables.
                    we can access local variable only inside the functions or local space

```python
x = 100

def func():
    x = 200
    print(x)

print(x)
func()
```

```python
x = 100

def func():
    x = 10
    x = x + 20
    print(x)

print(x)
func()
```

```python
x = 100

def func():
    x = x + 20
    print(x)

print(x)
func()
```

Global (Change inside local space) use "global" keyword

```python
x = 100

def func():
    global x
    x = x + 20
    print(x)

print(f"Before x: {x}")
func()
print(f"After x: {x}")
```
