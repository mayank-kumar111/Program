# Introduction to Python Programming

#### Data types
1. Numbers
    - int      --> complete numbers   (23,45,-44)
    - float    --> incomplete numbers  (-3.4, 5.5)
    - complex  --> real + imag.      (3+4j, 2.4-5j)
2. Str
    - collection of characters
    - ' ', " ", """ """
    - ordered
    - immutable (unChangeble Once unchangable once a  value is assigned to a variable, it cannot be changed. This is because variables in Python are immutable, meaning that they cannot be modified after they have been created. If you try to change the value of a variable, you will get an error. )
3. List
    - collection of objects (An Object can be anythings like str,char,num)
    - [ ] (Define By Square Brackets)
    - ordered
    - mutable (Changable)
4. Tuple
    - collection of objects
    - ( ) (Define By Round Brackets)
    - ordered
    - immutable
    - faster than list
5. Dict
    - collection of key & value pairs
    - { } (Define By Curley Brackets)
    - ordered
    - mutable
    - keys must be unique & immutable
6. Set (Are only Unordered)
    - collection of unique elements
    - { } (Define By Curley Brackets)
    - unordered
    - mutable
7. Bool
    - True
    - False
