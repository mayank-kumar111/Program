```python
# write some pandas code to create a dataframe from a dictionary
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
print(df)
```

```python
Functions in Pandas
1
2
3
Pandas is a powerful library in Python used for data manipulation and analysis. It provides numerous functions to handle, clean, and analyze data efficiently. Here are some of the most important functions in Pandas:

Data Loading and Inspection

read_csv(): This function is used to load data from a CSV file into a DataFrame. import pandas as pd df = pd.read_csv('data.csv')

head(): Returns the first n rows of a DataFrame (default is 5). df.head()

tail(): Returns the last n rows of a DataFrame (default is 5). df.tail()

info(): Provides a summary of the DataFrame, including the column names, data types, and non-null values. df.info()

describe(): Generates descriptive statistics of the DataFrame, such as mean, standard deviation, min, and max. df.describe()

Data Cleaning

isnull(): Detects missing values in the DataFrame. df.isnull()

fillna(): Fills missing values with a specified value. df.fillna(0)

dropna(): Drops rows or columns with missing values. df.dropna()

replace(): Replaces specified values with another value. df.replace('old_value', 'new_value')

Data Transformation

astype(): Casts a pandas object to a specified data type. df['column'] = df['column'].astype('int')

apply(): Applies a function along an axis of the DataFrame. df['column'] = df['column'].apply(lambda x: x * 2)

groupby(): Groups data by a specified column and performs aggregate functions. df.groupby('column').sum()

merge(): Merges two DataFrames based on a key column. df_merged = pd.merge(df1, df2, on='key_column')

Data Analysis

corr(): Computes the correlation between columns in the DataFrame. df.corr()

value_counts(): Returns the counts of unique values in a column. df['column'].value_counts()

pivot_table(): Creates a spreadsheet-style pivot table as a DataFrame. df.pivot_table(values='value_column', index='index_column', columns='column_to_pivot')

Data Visualization

plot(): Plots the data in the DataFrame. df.plot()

These functions are essential for data analysis and manipulation in Pandas. They help in loading, cleaning, transforming, and analyzing data efficiently.
```
