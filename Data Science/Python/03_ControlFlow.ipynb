### Conditional statement

1. #### if statement
                    syntax:

                            if condition:
                                st

```python
age = int(input("Enter age: "))
if age >= 18:
    print("you can vote")
```

```python
age = int(input("Enter age: "))
if age >= 18:    
    print("you can vote")
print("bye")
```

2. #### if else statement

                syntax:
                        if condition:
                            st
                        else:
                            st

```python
age = int(input("Enter age: "))
if age >= 18:
    print("you can vote")
else:
    print("you can't vote")
```

```python
num = int(input("Enter number: "))
if num > 0:
    print(f"{num} is +ve")
else:
    print(f"{num} is -ve")
```

```python
num = int(input("Enter num: "))
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

3. #### if elif else statement

                        syntax:
                                if condition:
                                    st
                                elif condition:
                                    st
                                elif condition:
                                    st
                                .
                                .
                                .
                                else:
                                    st

```python
day = int(input("Enter day number: "))

if day == 1:
    print("Mon")
elif day == 2:
    print("Tues")
elif day == 3:
    print('Wed')
elif day == 4:
    print("Thurs")
elif day == 5:
    print("Fri")
elif day == 6:
    print("Sat")
elif day == 7:
    print("Sun")
else:
    print("Invalid input...")
```

```python
data = [23, 45 , 67, 89, 12, 34, 56, 687, 90, 123, 456, 789]
i = 0
while i < len(data):
    if data[i] % 2 != 0:
        print(f"{data[i]} is odd")
    i += 1
```

```python
data = [23, 45 , 67, 89, 12, 34, 56, 687, 90, 123, 456, 789]
i = 0
while i < len(data):
    if data[i] % 2 == 0:
        print(f"{data[i]} is even")
    else:
        print(f"{data[i]} is odd")
    i += 1
```

```python
name = ["Mayank", "Rohit", "Suresh", "Anil", "Sunil", "Ramesh", "Suresh", "Anil", "Sunil", "Ramesh"]
i = 0
while i < len(name):
    if name[i].startswith("S"):
        print(name[i])
    i += 1
```

```python
i = 0
while i < len(name):
    if len(name[i]) < 5:
        print(name[i])
    i += 1
```

### Prime Number

```python
num = int(input("Enter num: "))
i = 2
while i < num:
    if num % i == 0:
        print(f"{num} is not prime")
        break
    i += 1
else:
    print(f"{num} is prime")
```
