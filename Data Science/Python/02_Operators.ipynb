Bitwise
1. Bitwise &
2. Bitwise |
3. Bitwise ~ (not)
4. Bitwise ^ (xor)
5. left shift
6. right shift

```python
10 & 8 # The bitwise AND operator (&) compares each bit of the first operand to the corresponding bit of the second operand. If both bits are 1, the corresponding result bit is set to 1. Otherwise, it is set to 0. In this case, the binary representation of 10 is 1010 and the binary representation of 8 is 1000. The bitwise AND of these two numbers will be 1000, which is equal to 8 in decimal. Therefore, the result of 10 & 8 will be 8.
```

```python
10 | 8
```

Bitwise NOT operation is represented by the ~ operator.

What it does

It inverts all the bits of a number:

Every 1 becomes 0
Every 0 becomes 1

```python
x = 5        # binary: 0000 0101
result = ~x  # binary: 1111 1010

print(result)  # Output: -6
```

To get the result of Bitwise NOT operation we use this formula:
1. x = -(x+1)

```python
10 ^ 8 # The bitwise XOR operator (^) compares each bit of the first operand to the corresponding bit of the second operand. If the bits are different, the corresponding result bit is set to 1. Otherwise, it is set to 0. In this case, the binary representation of 10 is 1010 and the binary representation of 8 is 1000. The bitwise XOR of these two numbers will be 0010, which is equal to 2 in decimal. Therefore, the result of 10 ^ 8 will be 2. 
```

1. Left Shift (<<)
2. Rule: x<<n=x * 2^n

3. Right Shift (>>)
4. Rule: x >> n = [x/2^n]

```python
10 << 2 # The left shift operator (<<) shifts the bits of the first operand to the left by the number of positions specified by the second operand. In this case, shifting the bits of 10 (which is 1010 in binary) to the left by 2 positions will result in 101000, which is equal to 40 in decimal. Therefore, the result of 10 << 2 will be 40.
```

```python
10 >> 2 # The right shift operator (>>) shifts the bits of the first operand to the right by the number of positions specified by the second operand. In this case, shifting the bits of 10 (which is 1010 in binary) to the right by 2 positions will result in 10, which is equal to 2 in decimal. Therefore, the result of 10 >> 2 will be 2.
```
