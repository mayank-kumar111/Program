### handling the missing values
#### there are 2 ways to handle the missing data
1. fill the missing values using central tendency / center of data
2. drop / delete the missing values. droping the data is not a good idea --> data loss

```python
df
```

```python
df.isnull()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df
```

```python
df['height'].loc[::2] = np.nan
df['age'].loc[::4] = np.nan
```

```python
df
```

```python
df.isna().sum()
```

```python
df.dropna(axis=0, how='any')
```

```python
df
```

```python
df.dropna(axis=0, how='all')
```

```python
df.loc['z'] = np.nan
```
