```python
# OTP Generator
 
import random
import string
def generate_otp(length=6):
    """Generate a random OTP of specified length."""
    characters = string.digits  # You can also include letters if needed
    otp = ''.join(random.choice(characters) for _ in range(length))
    return otp

if __name__ == "__main__":
    otp_length = 6  # You can change this to generate OTPs of different lengths
    otp = generate_otp(otp_length)
    print(f"Generated OTP: {otp}")
```

```python
# OTP Generator with Alphanumeric Characters
import random
import string
def generate_alphanumeric_otp(length=6):
    """Generate a random alphanumeric OTP of specified length."""
    characters = string.ascii_letters + string.digits  # Include both letters and digits
    otp = ''.join(random.choice(characters) for _ in range(length))
    return otp

if __name__ == "__main__":
    otp_length = 6  # You can change this to generate OTPs of different lengths
    otp = generate_alphanumeric_otp(otp_length)
    print(f"Generated Alphanumeric OTP: {otp}")

# OTP Generator with Custom Character Set
import random
def generate_custom_otp(length=6, character_set=string.digits):
    """Generate a random OTP of specified length using a custom character set."""
    otp = ''.join(random.choice(character_set) for _ in range(length))
    return otp

if __name__ == "__main__":
    otp_length = 6  # You can change this to generate OTPs of different lengths
    custom_characters = string.ascii_letters  # Example: Only letters
    otp = generate_custom_otp(otp_length, custom_characters)
    print(f"Generated Custom OTP: {otp}")
```
