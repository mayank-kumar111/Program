# Flipkart

```python
import requests
```

```python
url = "https://www.flipkart.com/search?q=motorola+mobile+5g&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_2_na_na_na&as-pos=2&as-type=RECENT&suggestionId=motorola+mobile+5g%7CMobiles&requestId=1bf01223-0f49-42e2-8fb4-168d370869a6&as-searchtext=motorola%20mobile%205g"
```

```python
resp = requests.get(url)
resp
```

```python
resp.reason
```

```python
data = resp.content
data
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data)
soup
```

### Find all mobile name

```python
# Find all mobile names in the search result page. We can inspect the HTML structure of the page to find the appropriate tags and classes that contain the mobile names. In this case, we can see that the mobile names are contained in div tags with class "RG5Slk".
all_divs = soup.find_all("div", class_="RG5Slk")
all_divs
```

```python
len(all_divs)
```

```python
v = all_divs[0:24] # Get the first 24 mobile names from the list of all divs.
v
```

```python
len(v)
```

```python
v[0]
```

```python
v[0].text
```

```python
for i in v:
    print(i.text)
    print()
```

```python
mobile = []
for i in v:
    mobile.append(i.text)
```

```python
print(mobile)
```

### Find Rating of mobiles

```python
all_divs = soup.find_all("div", attrs={'class':'MKiFS6'})
all_divs
```

```python
len(all_divs)
```

```python
v = all_divs[0:24] # Get the first 24 mobile names from the list of all divs.
v
```

```python
len(v)
```

```python
v[0].text
```

```python
for i in v:
    print(i.text)
```

```python
ratings = []
for i in v:
    ratings.append(i.text)
```

```python
print(ratings)
```

### Find Prices Of mobiles
