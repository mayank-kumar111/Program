#### Spreadness of data
1. std
2. var
3. quantiles
4. IQR

```python
ages
```

```python
np.mean(ages)
```

```python
np.std(ages) # This will calculate the standard deviation of the ages array, which measures the amount of variation or dispersion of the ages from the mean.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(ages, 'ro')
plt.show()
```

```python
ages
```

```python
mean = np.mean(ages)
mean
```

```python
d = []
for i in ages:
    d.append(abs(mean - i)) # This will calculate the absolute difference between each age and the mean age, which is used to compute the standard deviation.
# abs is the absolute value function used to get the magnitude of the difference regardless of its sign
std_dev = np.sqrt(sum(d) / len(d)) # This will calculate the standard deviation manually by taking the square root of the average of the squared differences.
std_dev
```

```python
print(d)
```

```python
np.mean(d)
```

```python
normal = np.random.normal(70, 15, 1000)
normal
```

```python
np.mean(normal)
```

```python
np.std(normal)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(normal, bins=30, ec='k')
plt.show()
```

```python
d = np.quantile(normal, [.25, .50, .75])
d
```

```python
q1 = d[0]
q2 = d[1]
q3 = d[2]
```

```python
q1
```

```python
q2
```

```python
q3
```

```python
IQR = q3 - q1
IQR
```

```python
plt.boxplot(normal)
plt.show()
```

```python
normal
```

```python
q1
```

```python
q3
```

```python
IQR
```

```python
lower_bound = q1 - 1.5 * IQR # This will calculate the lower bound for identifying outliers in the normal distribution using the interquartile range (IQR). Any data point below this lower bound is considered an outlier.
lower_bound
```

```python
upper_bound = q3 + 1.5 * IQR # Calculate the upper bound for identifying outliers
upper_bound
```

```python
normal[(normal < lower_bound) | (normal > upper_bound)] # This will return the values in the normal array that are considered outliers, which are those that fall below the lower bound or above the upper bound calculated using the interquartile range (IQR).
```
