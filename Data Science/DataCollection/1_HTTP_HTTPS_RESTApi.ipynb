# API

```python
import requests
from dotenv import load_dotenv
import os

load_dotenv()  # Load environment variables from .env file
API_KEY = os.getenv("WEATHER_API_KEY", "").strip()
```

```python
lat = 26.922070
lon = 75.778885
```

```python
url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_KEY}"
url
```

```python
resp = requests.get(url)
resp
```

```python
resp.content
```

```python
data = resp.json()
data
```

```python
city_name = input("Enter city name: ")
url = f"https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={API_KEY}"
resp = requests.get(url)
d = resp.json()
current_temp = d['main']['temp']
min_temp = d['main']['temp_min']
max_temp = d['main']['temp_max']
print(f"current_temp: {current_temp}")
print(f"Min_temp: {min_temp}")
print(f"Max_temp: {max_temp}")
```

```python
city_name = input("Enter city name: ")
url = f"https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={API_KEY}&units=metric"
resp = requests.get(url)
d = resp.json()
current_temp = d['main']['temp']
min_temp = d['main']['temp_min']
max_temp = d['main']['temp_max']
print(f"current_temp: {current_temp}")
print(f"Min_temp: {min_temp}")
print(f"Max_temp: {max_temp}")
```

```python
def get_temp(city_name, units='standard'):
    """city_name --> str
       units --> by default standard
       options are metric and imperial"""
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={API_KEY}&units={units}"
    resp = requests.get(url)
    d = resp.json()
    current_temp = d['main']['temp']
    min_temp = d['main']['temp_min']
    max_temp = d['main']['temp_max']
    print(f"current_temp: {current_temp}")
    print(f"Min_temp: {min_temp}")
    print(f"Max_temp: {max_temp}")
```

```python
get_temp('pune')
```

```python
get_temp('pune', units='imperial')
```

```python
get_temp('pune', units='metric')
```
