```python
# Task 1 : Print the average marks of 5 students in a class. The marks are given as input by the user.
marks = []
for i in range(5):
    mark = float(input(f"Enter marks for student {i + 1}: "))
    marks.append(mark)
average = sum(marks) / len(marks)
print(f"The average marks of the 5 students is: {average}")
```

```python
# Task 1 : Print the average marks of 5 students in a class. The marks are given as input by the user using a class.
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def get_average(self):
        return sum(self.marks) / len(self.marks)

# Create student objects and calculate average
students = []
for i in range(5):
    name = input(f"Enter name for student {i + 1}: ")
    marks = []
    for j in range(3):  # Assuming 3 subjects
        mark = float(input(f"Enter marks for subject {j + 1} of {name}: "))
        marks.append(mark)
    student = Student(name, marks)
    students.append(student)

# Calculate and print the average marks for each student
for student in students:
    print(f"The average marks of {student.name} is: {student.get_average()}")
```

```python
# Task 2 : To find the number of digits in a number after decimal point.
number = float(input("Enter a number: "))
decimal_part = str(number).split('.')[1]  # Get the part after the decimal point
digit_count = len(decimal_part)
print(f"The number of digits after the decimal point is: {digit_count}")
```

```python
# Task 3 : Radom OTP generator using class.
import random
import string
class OTPGenerator:
    def __init__(self, length=6):
        self.length = length

    def generate_otp(self):
        characters = string.digits + string.ascii_letters
        otp = ''.join(random.choice(characters) for _ in range(self.length))
        return otp
    
otp_generator = OTPGenerator()
otp = otp_generator.generate_otp()
print(f"Generated OTP: {otp}")
```
