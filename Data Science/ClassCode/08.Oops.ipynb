```python
class Student:

    def __init__(self, m1, m2, m3):
        self.m1 = m1         # instance variable
        self.m2 = m2
        self.m3 = m3

    def avg(self):
        return (self.m1 + self.m2 + self.m3) / 3
```

### Inheritance

```python
# single level inheritance

class A:

    def method_1(self):
        print("Method-1 is working")

class B(A):

    def method_2(self):
        print("Method-2 is working")
```

```python
b1 = B()
```

```python
b1.method_2()
```

```python
b1.method_1()
```

```python
# multi level inheritance

class A:

    def method_1(self):
        print("Method-1 is working")

class B(A):

    def method_2(self):
        print("Method-2 is working")

class C(B):

    def method_3(self):
        print("Method_3 is working")
```

```python
c1 = C()
```

```python
c1.method_2()
```

```python
# multiple inheritance

class A:

    def method_1(self):
        print("Method-1 is working")

class B:

    def method_2(self):
        print("Method-2 is working")

class C(A,B):

    def method_3(self):
        print("Method_3 is working")
```

```python
c1 = C()
```

```python
c1.method_2()
```

```python
# multiple inheritance

class A:

    def method_1(self):
        print("Method-1 is working")

    def show(self):
        print("This is show method in class A")

class B:

    def method_2(self):
        print("Method-2 is working")

    def show(self):
        print("This is show method in class B")

class C(A,B):         # MRO  --> method resolution order

    def method_3(self):
        print("Method_3 is working")
```

```python
# multiple inheritance

class A:

    def method_1(self):
        print("Method-1 is working")

    def show(self):
        print("This is show method in class A")

class B:

    def method_2(self):
        print("Method-2 is working")

    def show(self):
        print("This is show method in class B")

class C(A,B):         # MRO  --> method resolution order

    def method_3(self):
        super().show()
        B.show(self)
        print("Method_3 is working")
```

```python
class Person:

    def __init__(self,name,age):
        self.name = name
        self.age = age

    def details(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

    def __lt__(self,other):
        if self.age < other.age:
            return True
        else:
            return False
```

##### password validation

```python
password = input("Enter password: ")

if len(password) >= 8:
    lower = False
    upper = False
    digit = False
    special = False
    for i in password:
        if i.islower():
            lower = True
        elif i.isupper():
            upper = True
        elif i.isdigit():
            digit = True
        elif i in string.punctuation:
            special = True
    else:
        if all([lower, upper, digit, special]):
            print("Password is valid...")
        else:
            print("Password is invalid...")
else:
    print("Password is invalid...")
```

##### Create a Time class and initialize it with hours and minutes.    
1. Make a method addTime which should take two time object and add them. E.g.- (2 hour and 50 min)+(1 hr and 20 min) is (4 hr and 10 min)    
2. Make a method displayTime which should print the time    .
3. Make a method DisplayMinute which should display the total minutes in the Time. E.g.- (1 hr 2 min) should display 62 minute.


```python
class Time:

    def __init__(self, hour, min):
        self.hour = hour
        self.min = min

    def addTime(self, other):
        m1 = self.hour * 60 + self.min
        m2 = other.hour * 60 + other.min
        total = m1 + m2
        h = total // 60
        m = total % 60
        return f"{h} hr and {m} min"

    def displayTime(self):
        return f"{self.hour} hr and {self.min} min"

    def displayMinute(self):
        m = self.hour * 60 + self.min
        return f"{m} minute"
        
```

```python
t1 = Time(2,50)
t2 = Time(1,20)
```

```python
t1.addTime(t2)
```

```python
t1.displayTime()
```

```python
t2.displayMinute()
```

```python
class Person:

    def __init__(self,lname,fname,phone,email):
        self.lname = lname
        self.fname = fname
        self.phone = phone
        self.email = email

    def __str__(self):
        if type(self.email) == list:
            return f"{self.lname}, {self.fname} -- Phone Number: {self.phone} -- Email Address(es): {', '.join(self.email)}"
        else:
            return f"{self.lname}, {self.fname} -- Phone Number: {self.phone} -- Email Address(es): {self.email}"
```

```python
class AddressBook:

    def __init__(self):
        self.data = dict()

    def add_contact(self,contact):
        self.data[contact.phone] = contact

    def lookup_contact(self,lname,fname=None):
        if fname:
            for i in self.data.values():
                if i.lname == lname and i.fname == fname:
                    print(i)
        else:
            for i in self.data.values():
                if i.lname == lname:
                    print(i)
```

```python
a = AddressBook()
```

```python
a.add_contact(Ali)
a.add_contact(Sama)
a.add_contact(Emily)
a.add_contact(Wesley)
```

```python
a.lookup_contact("Kamil")
```
