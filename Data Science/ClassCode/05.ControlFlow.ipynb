#### nested if else statement

            syntax:

                    if condition:
                        st

                            st

### if elif else statement

            syntax:
                    if condition:
                        st
                    elif condition:
                        st
                    elif condition:
                        st
                    .
                    .
                    .
                    else:
                        st

```python
day = int(input("Enter day number: "))

if day == 1:
    print("Mon")
elif day == 2:
    print("Tues")
elif day == 3:
    print("Wed")
elif day == 4:
    print("Thurs")
elif day == 5:
    print("Fri")
elif day == 6:
    print("Sat")
elif day == 7:
    print("Sun")
else:
    print("Invalid input")
```
