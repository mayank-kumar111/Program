```python
normal = np.random.normal(70, 15, 1000)
normal
```

```python
np.mean(normal)
```

```python
np.std(normal)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(normal, bins=30, ec='k')
plt.show()
```

```python
d = np.quantile(normal, [.25, .50, .75])
d
```

```python
q1 = d[0]
q2 = d[1]
q3 = d[2]
```

```python
q1
```

```python
q2
```

```python
q3
```

```python
IQR = q3 - q1
IQR
```

```python
plt.boxplot(normal)
plt.show()
```

```python
normal
```

```python
q1
```

```python
q3
```

```python
IQR
```

```python
lower_bound = q1 - 1.5 * IQR
lower_bound
```

```python
upper_bound = q3 + 1.5 * IQR
upper_bound
```

```python
normal[(normal < lower_bound) | (normal > upper_bound)]
```
