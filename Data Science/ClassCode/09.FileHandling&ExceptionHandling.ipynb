```python
try:
    file = open(input("Enter file path: "))
    d = file.read()
    print(d)
    file.close()
except FileNotFoundError as e:
    print("file not exists...")
```

```python
while True:
    try:
        file = open(input("Enter file path: "))
    except FileNotFoundError as e:
        print("file not exists...")
    else:
        d = file.read()
        print(d)
        break
    finally:
        file.close()
```
